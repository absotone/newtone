```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline
```

## Loading the dataset

```python
df = pd.read_csv("../created_datasets/initial_df.csv")

df.head()
```

## Graphing the lenghts of the notes' lists

```python
df["notes_len"] = df["notes_list"].apply(lambda x : len(x))
```

```python
plt.plot(df["notes_len"])
```

```python
df["notes_len"].mean()
```

```python
min_thresh = df["notes_len"].mean() - df["notes_len"].std()/1.5
max_thresh = df["notes_len"].mean() + df["notes_len"].std()/1.5

new_df = df[df["notes_len"] >= min_thresh]
new_df = new_df[new_df["notes_len"] <= max_thresh]
```
